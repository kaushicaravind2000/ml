```python
import numpy as np
import matplotlib.pyplot as plt
from activation import Sigmoid

np.random.seed(0)
num_points = 100
X = np.random.randn(num_points, 2)
#check to see if X[n] satifies the > 2x+1
y = (X[:, 1] > 2 * X[:, 0] + 1).astype(np.float32)
X = np.hstack([np.ones((num_points, 1)), X])
weights = np.random.randn(3)

sigmoid = Sigmoid()

learning_rate = 0.1
num_iterations = 1000

# Training logistic regression using gradient descent
for i in range(num_iterations):
    z = np.dot(X, weights)
    predictions = sigmoid(z)
    # print(predictions.shape)
    errors = y - predictions
    gradient = np.dot(X.T, errors)
    weights += learning_rate * gradient

# Decision boundary visualization
x_values = np.linspace(-3, 3, 100)
y_values = -(weights[0] + weights[1] * x_values) / weights[2]

plt.figure(figsize=(10, 6))
plt.scatter(X[:, 1], X[:, 2], c=y, cmap='bwr', alpha=0.7)
plt.plot(x_values, y_values, label='Decision Boundary', color='black')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Binary Classification with Logistic Regression')
plt.legend()
plt.show()

z = np.linspace(-10, 10, 400)
sigmoid_output = sigmoid(z)
sigmoid_gradient = sigmoid.gradient(z)

fig, axs = plt.subplots(2, 1, figsize=(10, 8))

axs[0].plot(z, sigmoid_output, label='Sigmoid')
axs[0].set_title('Sigmoid Activation Function')
axs[0].set_xlabel('Input')
axs[0].set_ylabel('Output')
axs[0].grid(True)
axs[0].legend()

axs[1].plot(z, sigmoid_gradient, label='Sigmoid Gradient', color='orange')
axs[1].set_title('Gradient of the Sigmoid Function')
axs[1].set_xlabel('Input')
axs[1].set_ylabel('Gradient')
axs[1].grid(True)
axs[1].legend()

plt.tight_layout()
plt.show()
```
